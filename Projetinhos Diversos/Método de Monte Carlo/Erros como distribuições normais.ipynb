```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import time
```

```python
#-------------------------------------------------------------------------------
# Função para criar coordenadas de um quadrado de tamanho N x N
# Parâmetro:
#   sq_size (int): Tamanho do lado do quadrado
# Retorna:
#   numpy.array: Array com coordenadas (x,y) do quadrado, incluindo (0,0) na primeira posição
#-------------------------------------------------------------------------------
def cria_sq(sq_size):
    square = [(0,0)]
    
    # Gera todas as combinações inteiras de (x,y) de 1 até sq_size
    for coord_x in range(1, sq_size+1):
        for coord_y in range(1, sq_size+1):
            square = np.append(square, [[coord_x, coord_y]], axis=0)
    return square
```

```python
#-------------------------------------------------------------------------------
# Calcula a razão entre as áreas de dois quadrados usando Método de Monte Carlo
# Parâmetros:
#   interacao (int): Número de pontos aleatórios gerados
#   limite_maior (int): Tamanho do lado do quadrado maior
#   limite_menor (int): Tamanho do lado do quadrado menor
# Retorna:
#   float: Proporção de pontos dentro do quadrado menor em relação ao maior
#-------------------------------------------------------------------------------
def razao_Area(interacao, limite_maior, limite_menor):
    # Gera coordenadas aleatórias inteiras entre 1 e limite_maior (ambos inclusivos)
    x = np.random.randint(1, limite_maior+1, interacao)
    y = np.random.randint(1, limite_maior+1, interacao)
    
    # Conta pontos que caem dentro do quadrado menor (x e y <= limite_menor)
    dentro_quadrado = np.sum((x <= limite_menor) & (y <= limite_menor))
    return dentro_quadrado / interacao
```

```python
t1 = time.time()

# Cria dois quadrados, um maior e outro menor
sq_maior = cria_sq(100)
sq_menor = cria_sq(80)
interacao = 10000  # Número de pontos por simulação

# Executa uma simulação e exibe o resultado
print(f"A razão entre os dois quadrados pelo método de Monte Carlo com {interacao} interações foi de {razao_Area(interacao,sq_maior[-1][0],sq_menor[-1][0])}")

t2 = time.time()
print(f"O tempo gasto pra executar o programa foi de {t2 - t1}segundos")
```

```python
#-------------------------------------------------------------------------------
# Validação Estatística
#-------------------------------------------------------------------------------
total_result = np.empty(0)
n = 10000

for i in range(n):
    resultado = razao_Area(interacao, sq_maior[-1][0], sq_menor[-1][0])
    total_result = np.append(total_result, resultado)  # Acumula resultados

print(total_result)  # Todas as razões obtidas após executar a simulação de 10000 iterações n vezes
```

```python
# Calculando a média e o desvio padrão dos resultados da simulação executada n vezes
media = np.mean(total_result)
desvio_padrao = np.std(total_result)

print("Média:", media)
print("Desvio Padrão:", desvio_padrao)
```

```python
# Gerando dados de uma distribuição normal com a média e desvio padrão obtidos
dados_normais = np.random.normal(media, desvio_padrao, 1000)

# Plotando os dados originais e a distribuição normal ajustada
plt.hist(total_result, bins=50, density=True, alpha=0.6, color='g', label='Dados Originais')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, media, desvio_padrao)
plt.plot(x, p, 'k', linewidth=2, label='Distribuição Normal Ajustada')
plt.legend()
plt.title('Dados Originais e Distribuição Normal Ajustada')
plt.show()
```
