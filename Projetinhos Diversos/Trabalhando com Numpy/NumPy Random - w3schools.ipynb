```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy import random
```

```python
#Generate a random integer from 0 to 100:
print(random.randint(low=10, high=100, size=(2,3,4)), '\n', '-'*20)

#Generate a random float from 0 to 1:
print(random.rand(2,3,4))
```

```python
"""Generate Random Number From Array
The choice() method allows you to generate a random value based on an array of values.
The choice() method takes an array as a parameter and randomly returns one of the values."""

x = random.choice([3, 5, 7, 9, 12, 15, 18, 21], size=(4, 2)) #distribuição uniforme
print(x,'\n')

"""We can generate random numbers based on defined probabilities using the choice() method of the random module.
The choice() method allows us to specify the probability for each value.
The probability is set by a number between 0 and 1, where 0 means that the value will never occur and 1 means that the value will always occur."""

#The sum of all probability numbers should be 1.
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(5, 10))
print(x)
```

```python
x = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))
y = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(2, 5))
z = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 2))
print(x, y, z, sep='\n\n\n')
```

```python
"""Random Permutations of Elements
A permutation refers to an arrangement of elements. e.g. [3, 2, 1] is a permutation of [1, 2, 3] and vice-versa.
The NumPy Random module provides two methods for this: shuffle() and permutation()"""

#The shuffle() method makes changes to the original array.
arr = np.array([1, 2, 3, 4, 5])

random.shuffle(arr)
print(random.shuffle(arr))
print(arr)


#The permutation() method returns a re-arranged array (and leaves the original array un-changed).
arr = np.array([1, 2, 3, 4, 5])
print(random.permutation(arr))
print(arr)
```

```python
"""Normal Distribution
loc - (Mean) where the peak of the bell exists.
scale - (Standard Deviation) how flat the graph distribution should be.
size - The shape of the returned array."""


x = random.normal(loc=1, scale=2, size=(2, 3))
print(x, '\n','-'*20)

y = random.normal(0, 1, size=(10,2))
print(y)
sns.displot(y, kind='kde')
```

```python
"""Binomial Distribution
Binomial Distribution is a Discrete Distribution.
It describes the outcome of binary scenarios, e.g. toss of a coin, it will either be head or tails.
It has three parameters:
n - number of trials.
p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).
size - The shape of the returned array."""

binomio = random.binomial(n=100, p=0.5, size=50)
print(binomio)
sns.displot(binomio, kind='kde')
sns.displot(binomio, kind='hist')
```

```python
"""Poisson Distribution
Poisson Distribution is a Discrete Distribution.
It estimates how many times an event can happen in a specified time. e.g. If someone eats twice a day what is the probability he will eat thrice?
It has two parameters:
lam - rate or known number of occurrences e.g. 2 for above problem.
size - The shape of the returned array."""

sns.displot(random.poisson(lam=5, size=100), kde=True)

plt.show()
```

```python
"""Uniform Distribution
Used to describe probability where every event has equal chances of occuring.
It has three parameters:
low - lower bound - default 0 .0.
high - upper bound - default 1.0.
size - The shape of the returned array."""

x = random.uniform(low=0.0, high=1.0, size=(10000))
sns.displot(x, kde=True)
```

```python
"""Logistic Distribution
Logistic Distribution is used to describe growth.
Used extensively in machine learning in logistic regression, neural networks etc.
It has three parameters:
loc - mean, where the peak is. Default 0.
scale - standard deviation, the flatness of distribution. Default 1.
size - The shape of the returned array."""


x = random.logistic(loc=1, scale=2, size=(2, 3))
print(x)
x = random.logistic(loc=1, scale=10, size=50000)
n = random.normal(loc=1, scale=10, size=50000)


sns.displot((x, n), legend=False, kind='ecdf' )
sns.displot((x, n), legend=False, kind='kde' )
plt.legend(('logística', 'normal'))
```

```python
"""Multinomial Distribution
Multinomial distribution is a generalization of binomial distribution.
It describes outcomes of multi-nomial scenarios unlike binomial where scenarios must be only one of two. e.g. Blood type of a population, dice roll outcome.
It has three parameters:
n - number of possible outcomes (e.g. 6 for dice roll). Número total de experimentos ou tentativas.
pvals - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).
size - The shape of the returned array.

Note: Multinomial samples will NOT produce a single value! They will produce one value for each pval.

Note: As they are generalization of binomial distribution their visual representation and similarity of normal distribution is same as that of multiple binomial distributions."""

x = random.multinomial(n=40, pvals=[1/6,1/6,1/6,1/6,1/6,1/6], size=100)
#print(x)

freq = np.sum(x, axis=0)
print(freq)
freq = freq.reshape(1,6)

sns.barplot(freq)
plt.show()
```

```python
# Parâmetros da distribuição multinomial
n = 1  # Número total de lançamentos por rodada
pvals = [1/6,2/6,1/12,1/24,1/6+1/24,1/6]  # Probabilidades iguais para cada face do dado
size = 10000  # Número de experimentos (ou rodadas)

# Gerar dados multinomiais
data = np.random.multinomial(n, pvals, size=size)
print(data)


# Somar a ocorrência de cada face em todas as rodadas
# Aqui, queremos ver quantas vezes cada face (1 a 6) aparece em geral
face_totals = np.sum(data, axis=0)

# Preparar os rótulos do eixo x para representar as faces do dado (1 a 6)
faces = np.arange(1, 7)

# Criar o gráfico de barras
sns.barplot(x=faces, y=face_totals, palette="viridis")
plt.xlabel("Faces do Dado")
plt.ylabel("Frequência Total de Ocorrência")
plt.title("Distribuição Multinomial para Lançamentos de Dados")
plt.show()
```

```python
"""Exponential Distribution
Exponential distribution is used for describing time till next event e.g. failure/success etc.
It has two parameters:
scale - inverse of rate (see lam in poisson distribution) defaults to 1.0. Quanto tempo entre um evento e outro
size - The shape of the returned array.
Ela é basicamente o inverso da Poisson

A distribuição exponencial é uma distribuição de probabilidade contínua que modela o tempo entre eventos em um processo de Poisson, onde os eventos ocorrem de forma totalmente aleatória e independente entre si. Em outras palavras, a distribuição exponencial é útil para descrever o tempo de espera até que um evento aconteça, como o tempo até a próxima falha de uma máquina, o tempo de atendimento ao próximo cliente, ou o tempo de espera até que um carro chegue a um pedágio."""

data = random.exponential(scale=60, size=1000) #um evento por minuto em "média"

sns.kdeplot(data, color="purple")
plt.xlabel("Tempo entre eventos")
plt.ylabel("Frequência")
plt.title("Distribuição Exponencial de Tempo entre Eventos")
plt.show()
```

```python
#Convertendo uma exponencial para normal pelo TLC

normal = random.normal(loc = 1, size=10000)
exp = random.exponential(scale=1,size=(2,2))
data = np.mean(exp, axis=(0))

print(data)



# Criando o gráfico de densidade de probabilidade
sns.kdeplot(np.sort(data), color="Tomato", fill=True)  # 'fill=True' preenche a área sob a curva
plt.legend(["TLC"])
plt.title("Aproximação da Distribuição Normal por Soma de Variáveis Exponenciais")
plt.xlabel("Valor")
plt.ylabel("Densidade de Probabilidade")
#plt.xlim(0, 5)  # Ajuste o limite do eixo x, se necessário
#plt.ylim(0, 0.5)  # Ajuste o limite do eixo y, se necessário
plt.grid()
plt.show()

sns.kdeplot((normal), color=("green"), legend=False)
plt.show()
```

```python
"""Chi Square Distribution
Chi Square distribution is used as a basis to verify the hypothesis.
It has two parameters:
df - (degree of freedom).
size - The shape of the returned array."""

sns.kdeplot(random.chisquare(df=9, size=1000), color="Tomato", fill=True, bw_adjust=0.5)
plt.xlabel("Valor")
plt.ylabel("Densidade de Probabilidade")
plt.grid()
plt.show()
```
