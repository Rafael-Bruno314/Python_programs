```python
import os
from pdf2docx import Converter

def convert_to_word(pdf_path, output_dir):
    """Converte PDF para Word e salva no diretório especificado"""
    # Cria o nome do arquivo de saída
    file_name = os.path.basename(pdf_path).replace(".pdf", ".docx")
    output_path = os.path.join(output_dir, file_name)
    
    # Converte PDF para Word
    try:
        cv = Converter(pdf_path)
        cv.convert(output_path, start=0, end=None)
        cv.close()
        print(f" {os.path.basename(pdf_path)} convertido com sucesso para {output_path}")
    except Exception as e:
        print(f" Erro ao converter {pdf_path}: {str(e)}")

# Diretórios
diretorio_pdf = r"C:\Users\Rafael Bruno\Documents\Programacao\Python\Converter PDF em Word"
diretorio_word = os.path.join(diretorio_pdf, "diretorio_word")

# Cria o diretório de saída se não existir
os.makedirs(diretorio_word, exist_ok=True)

# Processa todos os PDFs no diretório
for file_name in os.listdir(diretorio_pdf):
    if file_name.lower().endswith(".pdf"):
        pdf_path = os.path.join(diretorio_pdf, file_name)
        convert_to_word(pdf_path, diretorio_word)

print("Conversão concluída!")
```
