```python
"""What is a Series?
A Pandas Series is like a column in a table.

It is a one-dimensional array holding data of any type."""

#Create a simple Pandas Series from a list:

a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar)

#Return the first value of the Series:

print(myvar[0], '\n')

#With the index argument, you can name your own labels (index são a identificação das linhas).

myvar = pd.Series(a, index = ["x", "y", "z"])
print(myvar)
print(myvar["y"])

#You can also use a key/value object, like a dictionary, when creating a Series.

calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

```python
"""A Pandas DataFrame is a 2 dimensional data structure, like a 2 dimensional array, or a table with rows and columns."""

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)
print(df) 

#Pandas use the loc attribute to return one or more specified row(s)

print('\n', df.loc[0])

#Note: This example returns a Pandas Series.
print(type(df.loc[0]))

print('\n', df.loc[[0, 1]])
#Note: When using [], the result is a Pandas DataFrame.

print(type(df.loc[[0, 1]]))

#With the index argument, you can name your own indexes.

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df, '\n')

print(df.loc["day2"])
```

```python
"""Remove Rows
One way to deal with empty cells is to remove rows that contain empty cells."""

df = pd.read_csv('data.csv')

new_df = df.dropna()

print(new_df.head())

#If you want to change the original DataFrame, use the inplace = True argument

df.dropna(inplace = True)

print('\n', df.head(17))
```

```python
"""Replace Empty Values
Another way of dealing with empty cells is to insert a new value instead.

This way you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value:"""

#Replace NULL values in the "Calories" columns with the number 130:
df = pd.read_csv('data.csv')

print(df.loc[17])

#df["Calories"].fillna(130, inplace = True)

df.fillna({"Calories":130}, inplace=True)

print('\n', df.loc[17])
```

```python
"""Replace Using Mean, Median, or Mode
Pandas uses the mean() median() and mode() methods to calculate the respective values for a specified column"""

df = pd.read_csv('data.csv')
print(df.loc[17])

x = df["Calories"].mean()
print(f"Média caloria: {x:.2f}")

#x = df["Calories"].median()
#x = df["Calories"].mode()[0]

df.fillna({"Calories":x}, inplace=True)
print('\n', df.loc[17].astype(int))
```

```python
"""Finding Relationships
A great aspect of the Pandas module is the corr() method.

The corr() method calculates the relationship between each column in your data set."""

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_json('new_dirty_data.json')
correlation_matrix = df.corr()

print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.set_theme(style="white")  # estilo de fundo




# Plotar o mapa de calor da matriz de correlação
sns.heatmap(
    correlation_matrix,
    annot=True,          # Adiciona os valores das correlações nas células
    fmt=".2f",           # Formato de duas casas decimais para os valores
    cmap="coolwarm",     # Escolhe o esquema de cores (aqui um gradiente de cores frias/quentes)
    vmin=-1, vmax=1,     # Define o limite mínimo e máximo das cores
    linewidths=.5,       # Define a largura das linhas entre as células
    square=True,         # Configura o gráfico como quadrado para uniformidade
    cbar_kws={"shrink": .8}  # Ajusta o tamanho da barra de cores
)

# Adicionar título e layout
plt.title("Matriz de Correlação", fontsize=16)
plt.xticks(rotation=45, ha='right')  # Rotaciona os rótulos do eixo x para melhor visualização
plt.yticks(rotation=0)  # Define os rótulos do eixo y na horizontal
plt.tight_layout()      # Ajusta o layout para que tudo caiba na figura
plt.show()

#--------------------

# Mapa de calor com clusters (hierarquia de similaridade)
sns.clustermap(
    correlation_matrix, 
    annot=True, 
    fmt=".2f", 
    cmap="coolwarm", 
    vmin=-1, vmax=1, 
    linewidths=.5, 
    figsize=(10, 8),
    cbar_kws={"shrink": .8}
)
plt.title("ClusterMap da Matriz de Correlação", fontsize=16)
plt.show()


#---------------------

# Gráfico de pares com linha de regressão
sns.pairplot(correlation_matrix, kind="reg", plot_kws={'line_kws':{'color':'red'}, 'scatter_kws': {'alpha':0.6}})
plt.suptitle("Pairplot com Linhas de Regressão", y=1.02, fontsize=16)
plt.show()


#----------------------

# Máscara para a metade superior da matriz
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Cria a figura e o heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(
    correlation_matrix, 
    mask=mask,             # Aplica a máscara
    annot=True,            # Mostra os valores
    fmt=".2f", 
    cmap="coolwarm", 
    vmin=-1, vmax=1, 
    linewidths=.5, 
    cbar_kws={"shrink": .8}
)

# Adiciona marcadores circulares nos pontos de correlação
for i in range(correlation_matrix.shape[0]):
    for j in range(i+1, correlation_matrix.shape[1]):
        plt.scatter(j + 0.5, i + 0.5, s=abs(correlation_matrix.iat[i, j]) * 300, color="black", alpha=0.5)

plt.title("Matrixplot com Marcadores", fontsize=16)
plt.show()
```
